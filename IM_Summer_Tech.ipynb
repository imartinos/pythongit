# IM Summer Tech Portfolio Analysis

```python
import datetime

import pandas as pd
import pandas.io.data
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rc('figure', figsize=(8, 7))
```

```python
StockSymbols=['MSFT', 'LNKD','DATA', 'GE']
df = pd.io.data.get_data_yahoo(StockSymbols, 
                               start=datetime.datetime(2014, 7, 1), 
                               end=datetime.date.today())['Adj Close']
df.tail()
```

```python
rets = df.pct_change()
rets.tail()
```

```python
pd.scatter_matrix(rets, diagonal='kde', figsize=(10, 10));
```

```python
corr = rets.corr()
corr
```

```python
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns);
```

One thing we are often interested in is the relationship of expected returns and the risk we are taking one. Often there is a trade-off between the two.

Here we use `plt.annotate` to put labels on the scatterplot.

```python
plt.scatter(rets.mean(), rets.std())
plt.xlabel('Expected returns')
plt.ylabel('Risk')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```
